```python
import os
import subprocess
import time
from pyngrok import ngrok

# 1. Set your Ngrok Token here
NGROK_TOKEN = "YOUR_NGROK_TOKEN_HERE"
ngrok.set_auth_token(NGROK_TOKEN)

# 2. Install Ollama (requires zstd)
!apt-get update && apt-get install -y zstd
!curl -fsSL https://ollama.com/install.sh | sh

# 3. Start Ollama server in the background
subprocess.Popen(["ollama", "serve"])
time.sleep(5) # Wait for server to start

# 4. Pull the model (Qwen 2.5 14B or GLM4)
!ollama pull qwen2.5:14b

# 5. Start Ngrok tunnel to expose Ollama's API port (11434)
tcp_tunnel = ngrok.connect(11434)
public_url = tcp_tunnel.public_url
print("\n" + "="*50)
print(f"🚀 YOUR PUBLIC API URL: {public_url}/v1")
print("COPY THIS URL to your GitHub Actions Secrets as ZAI_API_URL")
print("="*50 + "\n")

# Keep the notebook running
try:
    while True:
        time.sleep(60)
except KeyboardInterrupt:
    print("Stopping server...")
```
